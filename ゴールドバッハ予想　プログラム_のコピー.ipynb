```python
min = 4 #ここには偶数を入れる
max = 302
odd_prime = []
for i in range(3,max+1,2):
    for j in range(3,i,2):
        if i % j ==0:
            break
    else:
        odd_prime.append(i)
print(odd_prime,len(odd_prime))

for num in range(min,max+1,2):
    print(num)
    i=0
    while odd_prime[i]<num:
        print('(',odd_prime[i],',',num-odd_prime[i],')',end='')
        i+=1
    print('')
```

```python
import random

def is_prime(n):
    if n == 2: return True
    if n == 1 or n & 1 == 0: return False

    d = (n - 1) >> 1
    while d & 1 == 0:
        d >>= 1

    for k in range(100):
        a = random.randint(1, n - 1)
        t = d
        y = pow(a, t, n)

        while t != n - 1 and y != 1 and y != n - 1:
            y = (y * y) % n
            t <<= 1

        if y != n - 1 and t & 1 == 0:
            return False

    return True

if __name__ == '__main__':
    print(is_prime(3))

```
