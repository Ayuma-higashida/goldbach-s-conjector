```python
#xが奇素数・yが正の奇数(2x+y)
min = 9 #ここには奇数を入れるよ
max = 30
odd_prime = []
for i in range(3,max+1,2):
    for j in range(3,i,2):
        if i % j ==0:
            break
    else:
        odd_prime.append(i)
print(odd_prime,len(odd_prime))

for num in range(min,max+1,2):
    print(num)
    i=0
    while odd_prime[i]<num:
        if (num-2*odd_prime[i])>0:
            print('(',odd_prime[i],',',num-2*odd_prime[i],')',end='')
        i+=1
    print('')
```

```python
#xが奇素数・yが正の奇数(x+2y)
min = 3
max = 45
odd_prime = []
for i in range (3,max+1,2):
    for j in range(3, i, 2):
        if i % j ==0:
            break
    else:
        odd_prime.append(i)
print(odd_prime,len(odd_prime))
for num in range(min,max+1,2):
    print(num)
    i=0
    while odd_prime[i]<num:
        if (num-odd_prime[i])>0:
          if (num-odd_prime[i])%4==2:
            print('(',odd_prime[i],',',num-2*odd_prime[i],')',end='')
        i+=1
    print('')    
```

```python
#余りの計算(左%右)
min = 9 #ここには奇数を入れる
max = 151
odd_prime = []
for i in range(3,max+1,2):
    for j in range(3,i,2):
        if i % j ==0:
            break
    else:
        odd_prime.append(i)
print(odd_prime,len(odd_prime))

for num in range(min,max+1,2):
    print(num)
    i=0
    while odd_prime[i]<num:
        if (num-2*odd_prime[i]) in odd_prime:
            print('|(',odd_prime[i]%(num-2*odd_prime[i]),')',end='')
        i+=1
    print('/')
```
